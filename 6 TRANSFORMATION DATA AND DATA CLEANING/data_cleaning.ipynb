### Cleaning columns with a combination of str method with NumPy

```python
color = df['color']
black = color.str.contains('black')
black[:5]
```

```python
white = color.str.contains('white')
```

### Clean the entire dataset with applymap function

```python
aac_shelter_cat_outcome_eng = []
with open("aac_shelter_cat_outcome_eng.txt") as file:
     for line in file:
         if '[orange]' in line:
             # Remember this `color` until the next is found
              color = line
         else:
             # keep `color` as last-seen
             aac_shelter_cat_outcome_eng.append((color, line))

aac_shelter_cat_outcome_eng[:5]
```

```python
shelter_cat = pd.DataFrame(aac_shelter_cat_outcome_eng,
                        columns=['color', 'age_upon_outcome'])

shelter_cat.head()
```

```python
def color(item):
     if ' weeks' in item:
         return item[:item.find(' weeks')]
     elif 'months' in item:
         return item[:item.find('months')]
     else:
         return item
```

```python
shelter_cat =  shelter_cat.applymap(color)
```

```python
shelter_cat.head()
```

###### Here the word week and months will disappear
