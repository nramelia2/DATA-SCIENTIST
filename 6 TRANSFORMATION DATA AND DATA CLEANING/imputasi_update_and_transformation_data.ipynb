```python
# Import library

import matplotlib.pyplot as plt
import seaborn as sns
import missingno as mno
from sklearn import linear_model

%matplotlib inline
```

```python
data = pd.read_csv('phones_data.csv')
data.head(10)
```

```python
to_drop = ['popularity', 'best_price', 'sellers_amount', 'screen_size', 'brand_name', 'model_name', 'brand', 'release_date']
data.drop(to_drop, inplace=True, axis=1)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.loc[data["lowest_price"] == 0.0, "lowest_price"] = np.NAN
data.loc[data["highest_price"] == 0.0, "highest_price"] = np.NAN
data.loc[data["memory_size"] == 0.0, "memory_size"] = np.NAN
data.loc[data["battery_size"] == 0.0, "battery_size"] = np.NAN

data.isnull().sum()[2:6]
```

```python
# Make matriks
mno.matrix(data, figsize = (20, 6))
```

```python
# Using Regression to account for lost data
# Columns can be specified based on needs
missing_columns = ["lowest_price", "highest_price", "memory_size", "battery_size"]
```

```python
def random_imputation(df, feature):

    number_missing = data[feature].isnull().sum()
    observed_values = data.loc[df[feature].notnull(), feature]
    data.loc[data[feature].isnull(), feature + '_imp'] = np.random.choice(observed_values, number_missing, replace = True)
    
    return data
```

```python
for feature in missing_columns:
    data[feature + '_imp'] = data[feature]
    data = random_imputation(data, feature)
```

```python
# Deterministic Regression Imputation

deter_data = pd.DataFrame(columns = ["Det" + name for name in missing_columns])

for feature in missing_columns:
        
    deter_data["Det" + feature] = data[feature + "_imp"]
    parameters = list(set(data.columns) - set(missing_columns) - {feature + '_imp'})
    
    # Create a Linear Regression model to estimate lost data
    
    model = linear_model.LinearRegression()
    model.fit(X = data[parameters], y = data[feature + '_imp'])
    
    # observe that we store an index of data missing from the original data framework
    
    deter_data.loc[data[feature].isnull(), "Det" + feature] = model.predict(data[parameters])[data[feature].isnull()]
```

```python
# Make matriks
mno.matrix(deter_data, figsize = (20,5))
```

```python
# Make chart

sns.set()
fig, axes = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(8, 8)

for index, variable in enumerate(["memory_size", "battery_size"]):
    sns.histplot(data[variable].dropna(), kde = False, ax = axes[index, 0])
    sns.histplot(deter_data["Det" + variable], kde = False, ax = axes[index, 0], color = 'red')
    
    sns.boxplot(data = pd.concat([data[variable], deter_data["Det" + variable]], axis = 1),
                ax = axes[index, 1])
    
plt.tight_layout()
```

```python
# Display statistical calculation values for specific columns

pd.concat([data[["memory_size", "battery_size"]]], axis = 1).describe().T
```

### Trimming vs Winsorizing

```python
# Import library scipy

import numpy as np
from scipy.stats.mstats import winsorize
from scipy.stats.mstats import trima

# enter array contains 1 - 10

a = np.array([3.76, 4.00, 3.56, 3.57, 3.23, 3.89, 3.89, 3.23, 2.56, 2.00])
```

```python
# Winsorize will replace 10% of the lowest value and 20% of the high value

wins = winsorize(a, limits=[0.1, 0.2])
wins
```

```python
# Trima will cut the value by a certain limit
trims = trima(a, limits=(3.00,3.90))
print(trims)
```

### Scaling : Normalization

```python
# import pandas dan numpy
import pandas as pd
import numpy as np

scaling = data[['popularity', 'best_price']]
scaling2 = pd.DataFrame(scaling)

scaling2
```

```python
# Calculating the mean
means = scaling2.mean(axis = 0)

# Calculate max - min
max_min = scaling2.max(axis = 0) - scaling2.min(axis = 0)

# Apply transformation to data
train_scaled = (scaling2 - means) / max_min

train_scaled
```
