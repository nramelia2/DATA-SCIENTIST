## Search for pencil with Tukey's fences

```python
q1 = olim_tokyo_remove.quantile(0.25)
q3 = olim_tokyo_remove.quantile(0.75)
iqr = q3 - q1
iqr
```

## Search for pencil with Tukey's fences (2)

```python
# Handle Warning
import warnings
warnings.filterwarnings('ignore')

# Outlier Filter
# True = Pencilan, False = Bukan Pencilan.
olim_tokyo_align, iqr_new = olim_tokyo.align(iqr, axis=1, copy=False, join='outer')
outlier_filter = (olim_tokyo < q1 - 1.5 * iqr_new) | (olim_tokyo > q3 + 1.5 * iqr_new)
outlier_filter
```

##  Search for pencil with Tukey's fences (3)

```python
olim_tokyo[outlier_filter['Total']] \
  .loc[:, ['NOCCode', 'Total']] \
  .sort_values(by=['Total'], ascending=False) # ascending = small to big if (TRUE)
```
