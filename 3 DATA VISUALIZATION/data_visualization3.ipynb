### Latihan (18)
Tampilkan visualisasi Stacked Histogram dari fitur Champion dengan mengubah tipe datanya menjadi category (astype) dalam dataset copa_america

```python
new_name ={'Total Goal':'totalGoal'}
df.rename(columns=new_name, inplace=True)
df.head()
```

```python
# visualisasi stacked histogram
df['Champion'] = df['Champion'].astype('category')
#df.head()
list1=list()
mylabels=list()
for gen in df.Champion.cat.categories:
    list1.append(df[df.Champion==gen].totalGoal)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.title('Copa America')
plt.show()
```
